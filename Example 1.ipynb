```python
class Rectangle:
  def __init__(self, length, width ):
    self.length = length
    self.width = width

  def Perimeter(self):
    return 2* (self.length + self.width)

  def Area(self):
    return self.length * self.width

  def Display(self):
    print("Perimeter of Rectangle is: ", self.Perimeter())
    print("Area is: ", self. Area())

rectangle = Rectangle(10,5)

rectangle.Display()
rectangle.Rectangle()
```
